```python
from scipy import misc
import matplotlib.pyplot as plt
face=misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
dim = face.shape
new_pic = face[int(face.shape[0]/4) : int(face.shape[0] *3/4), int(face.shape[1]/4) : int(face.shape[1] *3/4)]
new_pic[new_pic>220]=200
new_pic[new_pic<50]=0
plt.imshow(new_pic, cmap=plt.cm.gray)
plt.show()
```

```python
# Standardization: putting different variables on the same scale.

np.random.seed(0)
A = np.random.randint(0, 100, [10, 5])
avg= [A[:, i].mean() for i in range(0,A.shape[1])]
std= [A[:, i].std() for i in range(0,A.shape[1])]
B = (A - avg)/std
```
