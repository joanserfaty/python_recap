```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2, 10)
y = x**2

plt.figure(figsize=(12, 8))
plt.plot(x,y, c='red',lw=3, ls='--', label='Pow2')
plt.plot(x,x**3, c='blue',lw=3, ls='--', label='Pow3')
plt.title('Figure 1')
plt.xlabel('Axis X')
plt.ylabel('Axis Y')
plt.legend()
plt.show()
#plt.savefig('figure.png')

plt.figure(figsize=(12, 8))
plt.subplot(2,1,1) #lines, columns, position
plt.plot(x, y, c='red')
plt.subplot(2,1,2)
plt.plot(x, y, c='blue')
```

```python
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

iris = load_iris()
x = iris.data
y = iris.target
ax = plt.axes(projection='3d')
ax.scatter(x[:,0], x[:, 1], x[:, 2], c=y)
```

```python
f = lambda x, y: np.sin(x) + np.cos(y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X,Y)

Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

```python
dataset = {f'experience{i}':np.random.randn(100) for i in range(4)}

def graph(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,8))
    for k, i in zip(dataset.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(dataset[k])
        plt.title(k)

graph(dataset)
```
